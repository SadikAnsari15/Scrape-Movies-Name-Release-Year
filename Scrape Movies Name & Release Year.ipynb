# Import Modules

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Request page source from URL

```python
url = "https://editorial.rottentomatoes.com/guide/best-netflix-movies-to-watch-right-now/"
```

```python
page = requests.get(url)
```

### page = requests.get(url): This line sends a GET request to the specified URL using the get method from the requests library. The response from the server, including the HTML content of the page, is stored in the page variable.

```python
page
```

```python
# Display the page content
page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

### 'html.parser': This specifies the parser that BeautifulSoup should use to parse the HTML content. In this case, it's using the HTML parser that comes with BeautifulSoup. The parser is responsible for turning the raw HTML into a parse tree, which makes it easier to navigate and extract information from the HTML.

```python
scrapped_Movie = soup.find_all('div', class_='article_movie_title')
scrapped_Movie
```

```python
#parse movie name
Movies = []
for movie in scrapped_Movie:
    Movies.append(movie.get_text().strip())
    
Movies
```

```python
MoviesName = [movie.split('(')[0] for movie in Movies]
MoviesName
```

```python
# parse movie release year
release_year = [int(movie.split('(')[-1].split(')')[0]) for movie in Movies]
release_year
```

### movie.split('(')[-1].split(')')[0]: This uses the split method to extract the portion of the movie title between the last opening parenthesis '(' and the first closing parenthesis ')'. This assumes that the release year is enclosed in parentheses at the end of each movie title.

```python
# create dataframe 
data = pd.DataFrame()
```

```python
# Add data into our DataFrame
data['MovieName']= MoviesName
data['ReleaseYear']= release_year
```

```python
data.head()
```

```python
# Save Data in localstorage
data.to_csv('Netflix top 100 Movies.csv', index=False)
```
